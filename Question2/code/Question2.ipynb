```python
# Import necessary libraries for data visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the student performance dataset
data = pd.read_csv("StudentsPerformance.csv")

# Visualization 1: Bar chart for gender distribution
sns.countplot(x="gender", data=data)
plt.title("Gender Distribution")
plt.savefig("ns2/gender_distribution.png")
plt.show()

# Visualization 2: Scatter plot for math score vs. reading score
plt.scatter(data["math_score"], data["reading_score"], alpha=0.5)
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.title("Math Score vs. Reading Score")
plt.savefig("ns2/math_vs_reading.png")
plt.show()

# Visualization 3: Histogram of math scores
plt.hist(data["math_score"], bins=20)
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.title("Distribution of Math Scores")
plt.savefig("ns2/math_score_distribution.png")
plt.show()

# Visualization 4: Box plot of writing scores by gender
sns.boxplot(x="gender", y="writing_score", data=data)
plt.title("Writing Score by Gender")
plt.savefig("ns2/writing_score_by_gender.png")
plt.show()

# Visualization 5: Bar chart for test preparation course completion
sns.countplot(x="test_preparation_course", data=data)
plt.title("Test Preparation Course Completion")
plt.savefig("ns2/test_prep_course_completion.png")
plt.show()
```
